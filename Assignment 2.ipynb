```python
import pandas as pd
import networkx as nx
import requests
```

```python
iad_url = "https://api.aviationapi.com/v1/preferred-routes/search?origin=IAD"
response_iad = requests.get(iad_url)
iad = response_iad.json()
iad_data = pd.DataFrame(iad)
iad_data
```

```python
bwi_url = "https://api.aviationapi.com/v1/preferred-routes/search?origin=BWI"
response_bwi = requests.get(bwi_url)
bwi = response_bwi.json()
bwi_data = pd.DataFrame(bwi)
bwi_data
```

```python
dca_url = "https://api.aviationapi.com/v1/preferred-routes/search?origin=DCA"
response_dca = requests.get(dca_url)
dca = response_dca.json()
dca_data = pd.DataFrame(dca)
dca_data
```

```python
origin = pd.concat([iad_data,dca_data,bwi_data])
origin
```

```python
g=nx.Graph()
```

```python
g=nx.from_pandas_edgelist(origin, "origin","destination")
```

```python
nx.write_graphml(g,"origins.graphml")
```
